```python
X=dfCon.drop('strength',axis=1)
Y=dfCon['strength']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X, Y, test_size=0.3, random_state=7)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(Xtrain,Ytrain)

print(model.score(Xtrain,Ytrain))
print(model.score(Xtest,Ytest))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
model_pipeline=Pipeline([('poly',PolynomialFeatures(degree=2, interaction_only=True)),
    ('reg', LinearRegression())
    
])
model_pipeline.fit(Xtrain,Ytrain)
print(model_pipeline.score(Xtrain,Ytrain))
print(model_pipeline.score(Xtest,Ytest))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
model_pipeline=Pipeline([('poly',PolynomialFeatures(degree=3, interaction_only=True)),
    ('reg', LinearRegression())
    
])
model_pipeline.fit(Xtrain,Ytrain)
print(model_pipeline.score(Xtrain,Ytrain))
print(model_pipeline.score(Xtest,Ytest))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
model_pipeline=Pipeline([('poly',PolynomialFeatures(degree=4, interaction_only=True)),
    ('reg', LinearRegression())
    
])
model_pipeline.fit(Xtrain,Ytrain)
print(model_pipeline.score(Xtrain,Ytrain))
print(model_pipeline.score(Xtest,Ytest))
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_absolute_error
k = KFold(random_state=7, n_splits=5, shuffle=True)
```

```python
# Linear Regression
## Since the dataset is small theefore using complete dataset for cv
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
pipeline = Pipeline([
    ('poly', PolynomialFeatures(degree=3, interaction_only=True)),
    ('scaler', StandardScaler()),
    ('reg', LinearRegression())
])

pipeline.fit(Xtrain,Ytrain)
```

```python
print(cross_val_score(pipeline, X, Y, cv=k).mean())  #on cross validation data for root absolute error
```

```python
print(cross_val_score(pipeline, X, Y, cv=k).std())
```

```python
print((-cross_val_score(pipeline, X, Y, cv=k, scoring='neg_mean_squared_error').mean())**0.5)
```

```python
# Decision Tree

from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=7)

model.fit(Xtrain, Ytrain)



print(cross_val_score(model, X, Y, cv=k).mean())
print(cross_val_score(model, X, Y, cv=k).std())
print((-cross_val_score(model, X, Y, cv=k, scoring='neg_mean_squared_error').mean())**0.5)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=7)

model.fit(Xtrain,Ytrain)


print(cross_val_score(model, X, Y, cv=k).mean())
print(cross_val_score(model, X, Y, cv=k).std())
print((-cross_val_score(model, X,Y, cv=k, scoring='neg_mean_squared_error').mean())**0.5)
```

```python
from sklearn.ensemble import BaggingRegressor

model = BaggingRegressor(random_state=7)

model.fit(Xtrain, Ytrain)

print(cross_val_score(model, X, Y, cv=k).mean())
print(cross_val_score(model, X, Y, cv=k).std())
print((-cross_val_score(model, X, Y, cv=k, scoring='neg_mean_squared_error').mean())**0.5)
```

```python
# Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor(random_state=7)

model.fit(Xtrain, Ytrain)


print(cross_val_score(model, X, Y, cv=k).mean())
print(cross_val_score(model, X, Y, cv=k).std())
print((-cross_val_score(model, X, Y, cv=k, scoring='neg_mean_squared_error').mean())**0.5)
```

Random forest is more accurate than Gradient Boosting
